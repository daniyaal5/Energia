```python
def groupDataByHours(data, intervals, method):
    """
    Groups and aggregates time-series data by specified hourly intervals.

    This function processes the given data by grouping it based on 'AnonymisedMPRN' and then
    aggregates the time interval columns using the specified method (e.g., mean, sum).
    The data is then averaged over the specified number of intervals to produce a summary
    dataframe with clear and concise time intervals.

    Parameters:
    - data (DataFrame): The input DataFrame containing time-series data, where rows are indexed by 'AnonymisedMPRN'.
    - intervals (int): The number of time intervals to group together for aggregation. This determines how the 
        columns are combined to form the hourly groups.
    - method (str): The aggregation method to apply to the grouped data (e.g., "mean", "sum").

    Returns:
    - DataFrame: A DataFrame with columns representing the aggregated data for each group of hours.
        The time intervals are merged for clarity and labeled in a human-readable format.
        
    """
    cluster_data = data.groupby("AnonymisedMPRN").agg(method).iloc[:, 1:-1]
    hour = 0
    mean_df = pd.DataFrame()
    for i in range(len(cluster_data.columns) // intervals):
        grouped_df = cluster_data.iloc[:, i*intervals:(i+1)*intervals].mean(axis = 1)
        if intervals % 2 == 0:
            mean_df[f'{int(i * intervals/2)} - {int((i * intervals/2 + intervals/2) - 1)}:59'] = grouped_df
        else:
            start_minute = '' if i % 2 == 0 else ':30'
            end_minute = ':29' if i % 2 == 0 else ':59'
            mean_df[f'{int(i * intervals/2)}{start_minute} - {int((i * intervals/2 + math.ceil(intervals/2)) - 1)}{end_minute}'] = grouped_df
        if i == (len(cluster_data.columns) // intervals) - 1:
            if len(cluster_data.columns) % intervals != 0:
                if intervals % 2 == 0:
                    mean_df[f'{int((i+1) * intervals/2)} - 23:59'] = cluster_data.iloc[:, (i+1)*intervals:].mean(axis = 1)
                else:
                    start_minute = '' if end_minute == ':59' else ':30'
                    mean_df[f'{int((i+1) * intervals/2)}{start_minute} - 23:59'] = cluster_data.iloc[:, (i+1)*intervals:].mean(axis = 1)
        
    return mean_df
```

```python
hours = 6
grouped_data = groupDataByHours(energia_data_copy, hours*2, "mean")
```

```python
# Visualizing with histograms
for col in grouped_data.columns:
    plt.figure(figsize=(8, 4))
    sns.histplot(grouped_data[col], kde=True)
    plt.title(f'Histogram of {col}')
    plt.show()

# Visualizing with Q-Q plots
for col in grouped_data.columns:
    plt.figure(figsize=(6, 6))
    stats.probplot(grouped_data[col], dist="norm", plot=plt)
    plt.title(f'Q-Q Plot of {col}')
    plt.show()
```
