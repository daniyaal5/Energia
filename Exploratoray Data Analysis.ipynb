```python
import pandas as pd
import numpy as np
import glob 
import os
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

## Excel Files Concatenation

```python
# Path for all the weekly excel files for the month of March
path = "C:/Users/daniy/Desktop/Energia"
joined_files = os.path.join(path, "March*.csv") 
  
# A list of all joined files is returned 
joined_list = glob.glob(joined_files) 

energia_data = pd.concat(map(pd.read_csv, joined_list), ignore_index=True)
```

# Mean of Each Time Interval

```python
plt.bar(data_measures.columns[2:], data_measures.iloc[1, 2:], color='skyblue')

# Rotate the x labels
plt.xticks(rotation=90)
plt.title('Mean of Electricity Consumption for Each Time Interval')
plt.xlabel('Time Interval')
plt.ylabel('Consumption')
plt.tight_layout()
plt.show()
```

```python
plt.plot(data_measures.iloc[1, 2:], label = "Mean")
plt.plot(data_measures.iloc[3, 2:], label = "Min")
plt.plot(data_measures.iloc[-1, 2:], label = "Max")
plt.xticks(rotation=90)
plt.title('Mean, Min, and Max of Electricity Consumption for Each Time Interval')
plt.xlabel('Time Interval')
plt.ylabel('Consumption')
plt.legend()
plt.tight_layout()
plt.show()
```

# Weekdays and Weekends

```python
# Convert the ReadingDate column from object to datetime64
energia_data_copy['ReadingDate'] = pd.to_datetime(energia_data_copy['ReadingDate'])
energia_data_copy.info()
```

```python
energia_data_copy['DayOfWeek'] = energia_data_copy['ReadingDate'].dt.dayofweek
energia_data_copy.iloc[:, 5:]
```

```python
weekend_data = energia_data_copy[(energia_data_copy['DayOfWeek'] == 5) | (energia_data_copy['DayOfWeek'] == 6)]
weekend_data.shape
```

```python
weekend_data.describe()
```

```python
weekday_data = energia_data_copy[(energia_data_copy['DayOfWeek'] != 5) & (energia_data_copy['DayOfWeek'] != 6)]
weekday_data.shape
```

```python
weekday_data.describe()
```

```python
plt.bar(weekend_data.describe().columns[2:-1], weekend_data.describe().iloc[1, 2:-1], color='skyblue', label = "Weekends")
plt.bar(weekday_data.describe().columns[2:-1], weekday_data.describe().iloc[1, 2:-1], color='orange', alpha = 0.7, label = "Weekdays")

# Rotate the x labels
plt.xticks(rotation=90)
plt.title('Mean of Electricity Consumption on Weekends and Weekdays')
plt.xlabel('Time Interval')
plt.ylabel('Consumption')
plt.tight_layout()
plt.legend()
plt.show()
```

```python
plt.plot(weekday_data.describe().iloc[1, 2:-1], label = "Mean on Weekdays")
plt.plot(weekday_data.describe().iloc[3, 2:-1], label = "Min on Weekdays")
plt.plot(weekday_data.describe().iloc[-1, 2:-1], label = "Max on Weekdays")
plt.plot(weekend_data.describe().iloc[1, 2:-1], label = "Mean on Weekends")
plt.plot(weekend_data.describe().iloc[3, 2:-1], label = "Min on Weekends")
plt.plot(weekend_data.describe().iloc[-1, 2:-1], label = "Max on Weekends")
plt.xticks(rotation=90)
plt.title('Mean, Min, and Max of Electricity Consumption on Weekdays and Weekends for Each Time Interval')
plt.xlabel('Time Interval')
plt.ylabel('Consumption')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
for i in energia_data_copy.columns[5:]:
    plt.boxplot(energia_data_copy.iloc[:, 5:-1][i])
    plt.title(i)
    plt.show()
```

```python
time_interval_columns = energia_data_copy.columns[5:-1]
time_interval_columns
```

```python
# Visualizing with histograms
for col in time_interval_columns:
    plt.figure(figsize=(8, 4))
    sns.histplot(energia_data_copy.iloc[:, 5:-1][col], kde=True)
    plt.title(f'Histogram of {col}')
    plt.show()

# Visualizing with Q-Q plots
for col in time_interval_columns:
    plt.figure(figsize=(6, 6))
    stats.probplot(energia_data_copy.iloc[:, 5:-1][col], dist="norm", plot=plt)
    plt.title(f'Q-Q Plot of {col}')
    plt.show()
```
